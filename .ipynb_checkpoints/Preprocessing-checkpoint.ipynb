```python
import numpy as np
import pandas as pd
import csv
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("./skill_builder_data.csv", encoding = "latin1")
print(df.columns)
```

```python
df1 = df[['user_id', 'correct', 'skill_id', 'skill_name']]
df2 = df1.dropna(axis=0)
```

```python
#df2 = df1.dropna(subset=['skill_name'])
```

```python
print(df2.isnull().sum())
```

```python
df2.skill_id.nunique()
```

```python
skill = df2.skill_name.unique()
skill_id = range(len(skill))
skill_id_map = dict(zip(skill, skill_id))
```

```python
skill_id_map
```

```python
df3 = df2[['user_id', 'correct', 'skill_name']]
df3['skill_id'] = df3['skill_name'].apply(lambda x: skill_id_map[x])
assess = df3[['user_id', 'skill_id', 'skill_name', 'correct']]
```

```python
assess.to_csv('./data/skill_data.csv')
```

```python
assess.nunique()
```

```python
# import seaborn as sns

# sns.displot(iparhai["difficulty_level"])
```

```python
REQUIRE_COLS = ['user_id','skill_id','correct']
user_ids = assess.user_id.unique()
tuples = []
for id in user_ids:
    df1 = assess[assess.user_id == id]
    df1 = df1[REQUIRE_COLS]
    problems = [int(pid) for pid in df1.skill_id]
    corrects = [int(corr) for corr in df1.correct]
    num_problems = len(problems)
    
    result = (num_problems, problems, corrects)
    tuples.append(result)                          # 1000 rows of users, 3 columns
    
    #print(num_problems)
    #print (problems)
    #print (corrects)
    #print (result)
    
```

```python
X = np.array(problems)
X.shape
```

```python
train, test = train_test_split(tuples, test_size=0.2)
```

```python
X = np.array(train)
X.shape
```
